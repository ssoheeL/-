### 2.5) DataFrame 행과 열 선택하기

```python
# 원본 데이터프레임에서 종가만 가지고와서 수익률 게산
# 주의사항 series로 넘기는게아니라 dataframe으로 넘겨야 나중에 오류가 없음
def return_pct_df(dataFrame):   
    df_close= dataFrame[['Close']]
    df_close['Return'] = df_close.pct_change().fillna(0)
    return df_close
end_pct_df = return_pct_df(lotte_che_df)
end_pct_df
```

### DataFrame 합치기 
- concat, merge, join

```python
# 1. 삼성데이터 불러오기
#sheet name을 적거나 sheet 순서를 적어도 됨
df_samsung = pd.read_excel('samsung_naver_excel.xlsx', sheet_name='samsung', index_col=0)
df_samsung

#2. 네이버 데이터 불러오기
df_naver  = pd.read_excel('samsung_naver_excel.xlsx', sheet_name =1, index_col=0)
df_naver.head()

#두 데이터의 컬럼 이름은 같지만 날짜(인덱스가) 다름
```

### concat을 사용해서 세로로 합치기

```python
#concat을 이용해서 세로로 합치기 

df = pd.concat([df_samsung, df_naver], keys= ['naver','samsung'])
df.head(10)
```

```python
df = pd.concat([df_samsung, df_naver], keys=['naver', 'samsung'], names =['stock','날짜'])
df.head(10)
df.index
# 이중 인덱스로 설정됨
```

```python
#이중 인덱스라 인덱스를 괄호로 묶어서 사용해야함
df.loc[('naver', '2018-01-02')]
```

```python
# 하나의 인덱스만 사용하면 에러가 나는것을 볼 수 있음
df.loc('2018-01-02')
```

### concat을 이용해서 가로로 합치기 (날짜기준)

```python
# concat시 axis=1로 했을 떄 
# 결과값 => 전체 인덱스에 대해서 없는 부분은 NaN으로 들어감
# 존재하는 영역만 값이 들어간 것을 볼 수 있음
df = pd.concat([df_samsung, df_naver], keys=['naver','samsung'], axis=1)
df.head(20)
```

### 두 데이터프레임 공통부분(교집합)합치기 join

```python
df = pd.concat([df_samsung, df_naver], keys=['naver','samsung'], axis=1, join='inner')
df
```

### pd.concat으로 세개의 데이터 프레임을 합칠 수 있을까?
